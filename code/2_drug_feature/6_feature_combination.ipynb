```python
def combined_two_feature(feature1, feature2):
    temp = np.hstack([feature1, feature2])
    print('Original shape: feature1 = {}, feature2 = {}, merged = {}'.format(feature1.shape, feature2.shape, temp.shape))
    assert temp.shape[0] == feature1.shape[0] == feature2.shape[0]
    assert temp.shape[1] == feature1.shape[1] + feature2.shape[1]
    temp = np.unique(temp, axis = 1) ##################此处会打乱顺序！！！之后的norm_idx全错啦！！！准确率比不norm下降啦！
    print('Drop repeated columns shape = ', temp.shape)
    return temp
```

```python
def save_combined_feature(pwd, filename, x, y, mf_shape_1 = False):
    if mf_shape_1: 
        col_index = np.array(range(mf_shape_1, x.shape[1]))
    else:
        col_index = np.array(range(x.shape[1]))
    np.savez_compressed(pwd + filename, 
                        x_t_np = x, 
                        y_t_np = y, 
                        norm_idx = col_index) # need_normalized_col_index
    print('# The columns needed to be normalized: ', col_index.shape)
    return
```

```python
def combined_process(target):
    pwd = '/home/chujunyi/2_Program/2_output_file/2_multilabel/1_X_Y_data/'
    morganR2_mfs = np.load(pwd + target + '_U_T_xy_np_rdkit_morgan_radius2.npz')
    word2vec = np.load(pwd + target + '_U_T_xy_np_word2vec.npz')
    combined_4mds = np.load(pwd + target + '_U_T_xy_np_combined_4mds.npz')

    x_t_np_morganR2_mfs, y_t_np_morganR2_mfs = np.unique(morganR2_mfs['x_t_np'], axis = 1), morganR2_mfs['y_t_np']
    x_t_np_word2vec, y_t_np_word2vec = np.unique(word2vec['x_t_np'], axis = 1), word2vec['y_t_np']
    x_t_np_combined_4mds, y_t_np_combined_4mds = np.unique(combined_4mds['x_t_np'], axis = 1), combined_4mds['y_t_np']
    
    assert (y_t_np_word2vec == y_t_np_morganR2_mfs).all()
    assert (y_t_np_word2vec == y_t_np_combined_4mds).all()
    
    print(x_t_np_word2vec.shape, x_t_np_morganR2_mfs.shape, x_t_np_combined_4mds.shape)
    print(x_t_np_morganR2_mfs.dtype, x_t_np_word2vec.dtype,x_t_np_combined_4mds.dtype)
    
    morganR2_word2vec = combined_two_feature(x_t_np_morganR2_mfs, x_t_np_word2vec) # 0
    word2vec_combined4MDs = combined_two_feature(x_t_np_word2vec, x_t_np_combined_4mds) # 0
    morganR2_combined4MDs = combined_two_feature(x_t_np_morganR2_mfs, x_t_np_combined_4mds) # 13
    morganR2_word2vec_combined4MDs = combined_two_feature(morganR2_word2vec, x_t_np_combined_4mds) # 13
    
    y = y_t_np_word2vec
    save_combined_feature(pwd, target + '_U_T_xy_np_morganR2_word2vec.npz', morganR2_word2vec, y, x_t_np_morganR2_mfs.shape[1])
    save_combined_feature(pwd, target + '_U_T_xy_np_word2vec_combined4MDs.npz', word2vec_combined4MDs, y, False)
    save_combined_feature(pwd, target + '_U_T_xy_np_morganR2_combined4MDs.npz', morganR2_combined4MDs, y, x_t_np_morganR2_mfs.shape[1])
    save_combined_feature(pwd, target + '_U_T_xy_np_morganR2_word2vec_combined4MDs.npz', morganR2_word2vec_combined4MDs, y, x_t_np_morganR2_mfs.shape[1])
    
    return
```
