```python
def read_fps_file(filename, first_line = 0):
    with open(filename, 'r') as f:
        fps = f.readlines()[first_line:]
    fps = [eval(item.replace('\n', '')) for item in fps]

    fps_list = []
    for item in fps:
        if isinstance(item, tuple):
            fps_list.append(sorted(list(item)))
        else:
            fps_list.append([item])

#     print(len(fps_list))
    return fps_list
```

```python
def read_bitvector_fps(filename, ex = ''):
    fps = pd.read_csv(filename, index_col = 0)
    fps_list =  [[eval(item[len(ex):]) for item in fps.iloc[i,:][fps.iloc[i,:] == 1].index.tolist()] for i in range(fps.shape[0])]
#     assert len(fps_list) == 580
    return fps_list
```

```python
def generate_fps_matrix(fps, shape_1, shape_2):
    mat = np.zeros((shape_1, shape_2))
    for idx, sample_fps in enumerate(fps):
        mat[idx, sample_fps] = 1
#     print(mat.shape)
    return mat
```

```python
def concat_split_fps(drug_data = gpcr_drug, target = 'GPCR', mf = 'FP2', mf_len = 1024):
    fps_list = []
    split_num = drug_data.shape[0] // 950 + 1
    mf_pwd = '/home/chujunyi/2_Program/2_output_file/3_drug_feature/3_MFs/split/'
    
    for i in range(split_num):

        l = i * 950
        r = (l+950) if (i < (split_num - 1)) else (l + drug_data.shape[0] % 950)

        mf_filename = target + '_updated_drug_' + mf + '_' + str(l) + '_' + str(r) + '.csv'

        if mf != 'PCP':
            fps_list += read_fps_file(mf_pwd + mf_filename, first_line = 0)
        else:
            fps_list += read_bitvector_fps(mf_pwd + mf_filename, ex = 'PubchemFP')
        
        print(mf_filename, len(fps_list))

    fps_mat = generate_fps_matrix(fps_list, drug_data.shape[0], mf_len)
    print(len(fps_list), fps_mat.shape)
    
    return fps_list, fps_mat
```

```python
def filter_columns(fps_np, print_ = True):
    fps = pd.DataFrame(fps_np)
    fps_drop0 = fps.loc[:, ~(fps == 0).all()]
    fps_drop1 = fps_drop0.loc[:, ~(fps_drop0 == 1).all()]
    fps_nondup = fps_drop1.T.drop_duplicates().T
    if print_: print('FPs shape: original = {}, filter0 = {}, filter1 = {}, drop_duplicate_column = {}'.format(
                     fps.shape, fps_drop0.shape, fps_drop1.shape, fps_nondup.shape))
    return fps_nondup
```

```python
def all_procedure_process_fps(fps_mat_np, drug_id_smiles = gpcr_drug, target = 'GPCR', mf = 'FP2'):
    
    fps_filtered = filter_columns(fps_mat_np)
    id_smi_fp = pd.concat([drug_id_smiles, fps_filtered], axis = 1)
    print(id_smi_fp.shape)
    
    save_pwd = '/home/chujunyi/2_Program/2_output_file/3_drug_feature/4_processed_MDs_MFs/'
    id_smi_fp.to_csv(save_pwd + target +'_U_' + mf + '_with_id.csv', index = False)
    return id_smi_fp
```

```python
def find_same_fps_id(fps_with_smi_id, dti_data):
    duplicate = fps_with_smi_id[fps_with_smi_id.iloc[:, 2:].duplicated(keep = False) == True]
    duplicate = duplicate.sort_values(list(duplicate.columns[2:]))
    print('Duplicate fps = {}, ratio dup/all = {}'.format(duplicate.shape, duplicate.shape[0]/fps_with_smi_id.shape[0]), end = ',')
    duplicate = duplicate.groupby(list(duplicate.columns[2:])).sum()
#     print(duplicate.columns)
    duplicate_dict = {}
    for idx, ids in enumerate(duplicate['drug_id']):
        ids = ids.replace('D', ',D')[1:].split(',')
        duplicate_dict[idx] = {}
        for id_ in ids:
            duplicate_dict[idx][id_] = list(dti_data[dti_data['drug_id'] == id_]['hsa_id'])
    print(len(duplicate_dict))
    return duplicate_dict
```

```python
def generate_dti_data_with_drug_feature(dti_data, drug_feature, target = 'GPCR', mf = 'FP2'):

    label_num = len(set(dti_data['hsa_id']))
    
    data_target_label = pd.concat([dti_data, pd.get_dummies(dti_data['hsa_id'])], axis = 1).groupby(by = 'drug_id').sum().reset_index()
    data_target_label = pd.merge(drug_feature, data_target_label, on = 'drug_id').sample(frac = 1.0, random_state = 1231).reset_index(drop = True)
    print(data_target_label.shape)
    x_target_label = data_target_label.iloc[:, 2:-label_num] # 2 columns:smiles, drug_id
    y_target_label = data_target_label.iloc[:, -label_num:]
   
    assert x_target_label.shape[1] == drug_feature.shape[1] - 2 # 2 columns:smiles, drug_id
    assert y_target_label.shape[1] == label_num
    print(x_target_label.shape, y_target_label.shape)
    x_t_np = np.array(x_target_label)
    y_t_np = np.array(y_target_label)
    print('X.shape = {}, Y.shape = {}'.format(x_t_np.shape, y_t_np.shape))
    
    drug_ids = np.array(data_target_label['drug_id'].tolist())
    target_ids = np.array(y_target_label.columns.tolist())
    assert (target_ids == data_target_label.columns[-len(target_ids):]).all()
    
    pwd = '/home/chujunyi/2_Program/2_output_file/2_multilabel/1_X_Y_data/'
    np.savez_compressed(pwd + target +'_U_T_xy_np_' + mf + '.npz', 
                        x_t_np = x_t_np, 
                        y_t_np = y_t_np,
                        norm_idx = [],
                        drug_ids = drug_ids,
                        target_ids = target_ids)
    return  drug_ids, target_ids, data_target_label, x_t_np, y_t_np
```

```python
def drop_unnumeric_columns(mds, print_ = True):
    if print_: print('Drop un_numeric columns: Original shape = {}'.format(mds.shape))
    if not isinstance(mds, pd.core.frame.DataFrame): mds = pd.DataFrame(mds)
    drop_column_list = []
    columns_name = mds.columns
    for i in range(2, mds.shape[1]): # 不更改id, smiles列
        num = 0
        for j in range(mds.shape[0]):
            if type(mds.iloc[j,i]) is str:
                try:
                    mds.iloc[j,i] = float(mds.iloc[j,i])
                except:
                    num += 1
                    mds.iloc[j,i] = np.nan
            if num > 30: 
                drop_column_list.append(columns_name[i])
                break
    if print_: print(drop_column_list)
    
    for col in drop_column_list:
        del mds[col]
    if print_: print('Droped shape = {}'.format(mds.shape))
    return 
```

```python
def drop_nan_columns(mds, print_ = True):
    if print_: print('Original: MDs shape = ', mds.shape)

    drop_nan_list = [col for col in mds.columns if mds[col].isnull().sum() >= 30]
    mds.drop(drop_nan_list, axis = 1, inplace = True)
    if print_: print('After drop NAN >= 550 columns, shape = ', mds.shape)

    drop_inf_list = [col for col in mds.columns[2:] if (np.isinf(mds[col].astype('float')).any() or (np.max(mds[col]) >= 9.978482375030003e+178)  or (np.max(mds[col]) == np.inf))]
    print(drop_inf_list)
    mds.drop(drop_inf_list, axis = 1, inplace = True)
    if print_: print('After drop Inf columns, shape = ', mds.shape)

    fillna_column_list = [col for col in mds.columns if 0 < mds[col].isnull().sum() < 550]
    for col in fillna_column_list:
        mds[col].fillna(mds[col].mean(), inplace = True)
    if print_: print('Final: after fill NAN columns, shape = ', mds.shape)
    return
# drop_inf_list = [col for col in mds.columns[1:] if ((col != 'smiles') and (np.isinf(mds[col].astype('float')).any() or (np.max(mds[col]) == 9.978482375030003e+178) or (np.max(mds[col]) == np.inf)))]
```

```python
def filter_columns(mds, print_ = True):
    if not isinstance(mds, pd.core.frame.DataFrame): mds = pd.DataFrame(mds)
    mds_drop0 = mds.loc[:, ~(mds == 0).all()]
    mds_drop1 = mds_drop0.loc[:, ~(mds_drop0 == 1).all()]
    mds_nondup = mds_drop1.T.drop_duplicates().T
    if print_: print('MDs shape: original = {}, filter0 = {}, filter1 = {}, drop_duplicate_column = {}'.format(
                     mds.shape, mds_drop0.shape, mds_drop1.shape, mds_nondup.shape))
    return mds_nondup
```

```python
def all_procedure_process_mds(drug_data = gpcr_drug, target = 'GPCR', md = 'RDK', split_n = 950):
    _, mds = concat_split_mds(drug_data, target, md, split_n)

    drop_unnumeric_columns(mds)
    
    drop_nan_columns(mds)
    mds_filtered = filter_columns(mds)
    print('Finally shape: ', mds_filtered.shape)
    
    pwd = '/home/chujunyi/2_Program/2_output_file/3_drug_feature/4_processed_MDs_MFs/'
    mds_filtered.to_csv(pwd + target + '_U_MD_' + md + '_with_smi_id.csv', index = False)
    
    return mds_filtered
```

```python
def generate_dti_data_with_drug_feature(dti_data, drug_feature, target = 'GPCR', md = 'FP2'):

    label_num = len(set(dti_data['hsa_id']))
    
    data_target_label = pd.concat([dti_data, pd.get_dummies(dti_data['hsa_id'])], axis = 1).groupby(by = 'drug_id').sum().reset_index()
    data_target_label = pd.merge(drug_feature, data_target_label, on = 'drug_id').sample(frac = 1.0, random_state = 1231).reset_index(drop = True)
    print(data_target_label.shape)
    x_target_label = data_target_label.iloc[:, 2:-label_num] # 2 columns:smiles, drug_id
    y_target_label = data_target_label.iloc[:, -label_num:]
   
    assert x_target_label.shape[1] == drug_feature.shape[1] - 2 # 2 columns:smiles, drug_id
    assert y_target_label.shape[1] == label_num
    print(x_target_label.shape, y_target_label.shape)
    x_t_np = np.array(x_target_label).astype('float64')
    y_t_np = np.array(y_target_label)
    print('X.shape = {}, Y.shape = {}'.format(x_t_np.shape, y_t_np.shape))
    
    drug_ids = np.array(data_target_label['drug_id'].tolist())
    target_ids = np.array(y_target_label.columns.tolist())
    assert (target_ids == data_target_label.columns[-len(target_ids):]).all()
    
    pwd = '/home/chujunyi/2_Program/2_output_file/2_multilabel/1_X_Y_data/'
    np.savez_compressed(pwd + target +'_U_T_xy_np_MD_' + md + '.npz', 
                        x_t_np = x_t_np, 
                        y_t_np = y_t_np,
                        norm_idx = np.array(range(x_t_np.shape[1])),
                        drug_ids = drug_ids,
                        target_ids = target_ids)
    return  drug_ids, target_ids, data_target_label, x_t_np, y_t_np
```
