```python
def generate_dti_data_with_drug_feature(dti_data, drug_feature):

    label_num = len(set(dti_data['hsa_id']))
    
    data_target_label = pd.concat([dti_data, pd.get_dummies(dti_data['hsa_id'])], axis = 1).groupby(by = 'drug_id').sum().reset_index()
    data_target_label = pd.merge(drug_feature, data_target_label, on = 'drug_id').sample(frac = 1.0, random_state = 1231).reset_index(drop = True)
    print(data_target_label.shape)
    x_target_label = data_target_label.iloc[:, 2:-label_num] # 2 columns:smiles, drug_id
    y_target_label = data_target_label.iloc[:, -label_num:]
    assert x_target_label.shape[1] == drug_feature.shape[1] - 2 # 2 columns:smiles, drug_id
    assert y_target_label.shape[1] == label_num
    print(x_target_label.shape, y_target_label.shape)
    x_t_np = np.array(x_target_label)
    y_t_np = np.array(y_target_label)
    print('X.shape = {}, Y.shape = {}'.format(x_t_np.shape, y_t_np.shape))
    
    return data_target_label, x_t_np, y_t_np
```

```python
_, x_t_np_rdkit_morgan, y_t_np_rdkit_morgan = generate_dti_data_with_drug_feature(dti_data, rdkit_morgan_with_id)
_, x_t_np_rdkit_morgan4, y_t_np_rdkit_morgan4 = generate_dti_data_with_drug_feature(dti_data, rdkit_morgan4_with_id)
_, x_t_np_rdkit_maccs, y_t_np_rdkit_maccs = generate_dti_data_with_drug_feature(dti_data, rdkit_maccs_with_id)
_, x_t_np_padel_pubchem, y_t_np_padel_pubchem = generate_dti_data_with_drug_feature(dti_data, padel_pubchem_with_id)
_, x_t_np_cdk_pubchem, y_t_np_cdk_pubchem = generate_dti_data_with_drug_feature(dti_data, cdk_pubchem_with_id)
_, x_t_np_pybel_fp2, y_t_np_pybel_fp2 = generate_dti_data_with_drug_feature(dti_data, pybel_fp2_with_id)
_, x_t_np_cdk_graph, y_t_np_cdk_graph = generate_dti_data_with_drug_feature(dti_data, cdk_graph_with_id)
```
